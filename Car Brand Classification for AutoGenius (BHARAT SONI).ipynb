### Step 1 : Import library


```python
import pandas as pd
```

### Step 2: Read Dataset

```python
df = pd.read_csv('https://github.com/YBIFoundation/MLProjectIdeas/raw/main/Car%20Brand%20Classification%20for%20AutoGenius.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

### Data Preprocessing (Label Encoding)

```python
df.columns
```

```python
# Check categories
df["Logo Shape"].value_counts()

df["Headlight Design"].value_counts()

df["Grille Size"].value_counts()

df["Dominant Color"].value_counts()

df["Car Brand"].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder

brand_encoder = LabelEncoder()

df["Car Brand"] = brand_encoder.fit_transform(df["Car Brand"])
```

### Step 3: Define y and X

```python
y = df["Car Brand"]
X = df[["Logo Shape","Headlight Design","Grille Size",
    "Dominant Color","Wheel Base (cm)","Body Length (cm)"]]
X = pd.get_dummies(X, dtype=int)
```

### Step 4 : Train Test Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=2529)
```

### Step 5 : Select Model

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(random_state=2529)
```

### Step 6 : Train Model

```python
model.fit(X_train, y_train)
```

### Step 7 : Prediction

```python
y_pred = model.predict(X_test)
y_pred
```

### Step 8 : Evaluate

```python
from sklearn.metrics import accuracy_score
print("Accuracy Score")
print(accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import classification_report
print("Classification Report")
print(classification_report(y_test, y_pred))
```
